#### Linked Lists & Arrays

Linked Lists & Arrays are concerned with four main aspects: accessing, searching, inserting, and deleting data. Arrays are fast for accessing data due to their fixed length and contiguous memory allocation, but can be slow for searching, inserting, and deleting, especially with large volumes of data. Dynamic data structures, such as arrays, can expand or shrink to conserve memory, with functions for inserting and removing data. A more efficient method for array expansion and contraction involves doubling the size when full and halving it when usage falls to 25% capacity. Linked structures, an alternative to arrays, consist of nodes containing data and memory pointers, allowing for flexible data access, insertion, and removal.

What are concerned with 4 things in a data structure:

```python
#linked list manual implemetation with OOO

class Node:
    def __init__(self,data):
        self.data = data #stores values
        self.next = None #pointer to next node

class LinkedList:
    def __init__(self):
        self.head = None #head of the list

    def append(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node #add at the end

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

        
```

```python
#example usage

linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)

linked_list.display()
```
